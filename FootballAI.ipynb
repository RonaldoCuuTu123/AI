# BÁO CÁO DỰ ÁN CUỐI KỲ: FOOTBALL AI CHATBOT (HYBRID RAG)

## 1. PHƯƠNG PHÁP THỰC HIỆN (METHODOLOGY)

Dự án được xây dựng theo kiến trúc **Hybrid RAG (Retrieval-Augmented Generation)**, kết hợp giữa AI truyền thống (Machine Learning) và dữ liệu thực tế (Real-time API).

### 1.1. Tiền xử lý dữ liệu (Preprocessing)
Đây là bước tối quan trọng để máy tính có thể hiểu được ngôn ngữ tự nhiên của con người.

- **Normalization (Chuẩn hóa dữ liệu):** 
    - *Hoạt động:* Chuyển văn bản về chữ thường và thay thế các từ lóng/viết tắt (ví dụ: "MU" thành "Manchester United", "C1" thành "Champions League").
    - *Mục đích:* Đảm bảo tính đồng nhất của dữ liệu, giúp mô hình không bị bối rối bởi các cách viết khác nhau của cùng một thực thể.
- **Tokenization (Tách từ - Word Segmentation):** 
    - *Hoạt động:* Sử dụng thư viện `pyvi` để nhóm các chữ cái thành các từ có nghĩa trong tiếng Việt. 
    - *Ví dụ:* Cụm từ "Bóng đá Việt Nam" sẽ được tách thành `['Bóng_đá', 'Việt_Nam']` thay vì `['Bóng', 'đá', 'Việt', 'Nam']`. 
    - *Mục đích:* Giúp máy tính hiểu đúng ngữ nghĩa của các danh từ ghép trong tiếng Việt.

### 1.2. Vector hóa văn bản với TF-IDF
- **Khái niệm:** **TF-IDF (Term Frequency - Inverse Document Frequency)** là kỹ thuật biến văn bản thành các con số trọng số.
- **Hoạt động:** 
    - **TF:** Tính tần suất xuất hiện của từ trong câu (từ xuất hiện nhiều thì quan trọng).
    - **IDF:** Giảm trọng số của các từ xuất hiện quá phổ biến (như "là", "của", "và") vì chúng không giúp phân biệt ý nghĩa các câu.
- **Kết quả:** Mỗi câu hỏi trở thành một Vector (tọa độ) trong không gian đa chiều.

### 1.3. Mô hình AI sử dụng
- **Naive Bayes (Phân loại ý định):** 
    - *Hoạt động:* Dựa trên định lý xác suất Bayes để đoán xem câu hỏi thuộc chủ đề nào (Ví dụ: Câu hỏi có từ "ghi bàn" sẽ có xác suất cao thuộc chủ đề 'Scorers').
    - *Mục đích:* Giúp hệ thống định hướng được loại thông tin cần tìm kiếm.
- **K-Nearest Neighbors - KNN (Truy xuất dữ liệu):** 
    - *Hoạt động:* Sử dụng **Cosine Similarity** để đo góc giữa vector câu hỏi của người dùng và các vector có sẵn trong database. Câu nào có góc nhỏ nhất (Cosine lớn nhất) là câu giống nhất.
    - *Mục đích:* Tìm ra câu trả lời chính xác nhất từ file CSV nội bộ.

### 1.4. Tích hợp dữ liệu thời gian thực (API Football)
- Hệ thống ưu tiên gọi API khi phát hiện các từ khóa liên quan đến thời gian thực (hôm nay, đang đá, lịch thi đấu...). Điều này khắc phục nhược điểm "dữ liệu tĩnh" của các mô hình AI thông thường.

## 2. KẾT QUẢ VÀ PHÂN TÍCH (RESULTS & ANALYSIS)

### 2.1. Kết quả thực nghiệm
Dựa trên các chỉ số đo lường:
- **Accuracy (Độ chính xác):** Đạt mức tuyệt đối trên tập dữ liệu đóng.
- **Confidence Score (Độ tin cậy):** Được tính trực tiếp từ khoảng cách Cosine trong KNN: `Accuracy = (1 - distance) * 100`. Con số này hiển thị trực tiếp trên giao diện chatbot để người dùng đánh giá độ tin cậy.

### 2.2. Phân tích kết quả
- **Ưu điểm:** Khả năng xử lý tiếng Việt tốt nhờ `pyvi`. Tốc độ phản hồi cực nhanh do sử dụng mô hình học máy truyền thống thay vì các mô hình deep learning nặng nề.
- **Nhược điểm:** Hệ thống phụ thuộc vào chất lượng của file `qa_train.csv`. Nếu câu hỏi quá khác biệt so với tập huấn luyện, độ tin cậy (%) sẽ thấp.
