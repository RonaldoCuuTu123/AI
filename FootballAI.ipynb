## 1. GIỚI THIỆU ĐỀ TÀI

Trong kỷ nguyên số, dữ liệu về bóng đá sinh ra với khối lượng khổng lồ. Việc tra cứu thủ công thường mất thời gian và thiếu tính tương tác. Dự án xây dựng một **Information Retrieval (IR) Chatbot** giúp người dùng tra cứu nhanh chóng kết quả, lịch thi đấu và thông tin bóng đá thông qua ngôn ngữ tự nhiên.

**Sự phát triển so với giai đoạn giữa kỳ:**
- Từ mô hình IR thuần túy dựa trên dữ liệu tĩnh (CSV), hệ thống đã phát triển thành kiến trúc **Hybrid RAG** (kết hợp dữ liệu nội bộ và Real-time API).
- Tối ưu hóa xử lý tiếng Việt chuyên sâu thay vì chỉ dùng NLTK cơ bản.
- Bổ sung tính năng đánh giá độ tin cậy của câu trả lời trực tiếp trên giao diện.

## 2. KIẾN TRÚC HỆ THỐNG VÀ PHƯƠNG PHÁP PHÁT TRIỂN

### 2.1. Tiền xử lý dữ liệu nâng cao (NLP Preprocessing)
- **Word Segmentation (Phát triển mới):** Thay thế NLTK bằng `pyvi` để tách từ tiếng Việt chính xác. Các cụm từ như "Ngoại hạng Anh" được xử lý thành `Ngoại_hạng_Anh` giúp giữ nguyên ngữ nghĩa.
- **Normalization:** Chuẩn hóa từ lóng và viết tắt (MU, EPL, C1, CR7...) giúp mô hình hiểu đúng thực thể bóng đá.

### 2.2. Biểu diễn ngôn ngữ (TF-IDF Vectorization)
- Sử dụng **TF-IDF Vectorizer** để biến văn bản thành vector trọng số. Kỹ thuật này giúp máy tính nhận diện được các từ khóa quan trọng trong câu hỏi bóng đá.

### 2.3. Mô hình xử lý Hybrid RAG
Hệ thống vận hành theo luồng xử lý kép được nâng cấp:
1. **Intent Classification (Naive Bayes):** Dự đoán chủ đề người dùng đang hỏi (Kết quả, Lịch thi đấu, Vua phá lưới...).
2. **Context Retrieval (KNN):** Tìm kiếm câu trả lời tương đồng nhất trong cơ sở dữ liệu dựa trên khoảng cách Cosine.
3. **Real-time API Integration (Phát triển mới):** Khi phát hiện ý định hỏi về dữ liệu thời gian thực (lịch thi đấu hôm nay, trực tiếp), hệ thống tự động gọi **Football API** để lấy dữ liệu mới nhất, khắc phục nhược điểm dữ liệu tĩnh của giai đoạn giữa kỳ.

## 3. KẾT QUẢ VÀ THỰC NGHIỆM

### 3.1. Đánh giá mô hình (Phát triển mới)
Chúng em đã triển khai script đánh giá chuyên sâu để đo lường hiệu năng:

| Mô hình | Chỉ số Accuracy | Mục tiêu |
| :--- | :--- | :--- |
| **Naive Bayes** | ~1.0000 | Phân loại đúng 100% các intent trong tập test nội bộ. |
| **KNN Retrieval** | ~1.0000 | Tìm được chính xác ngữ cảnh tương ứng với câu hỏi. |

### 3.2. Độ tin cậy (Confidence Score)
Hệ thống hiện tại đã có khả năng tính toán mức độ tin cậy dựa trên khoảng cách hình học:
$$Accuracy = (1 - Distance_{Cosine}) \times 100$$
Con số này được hiển thị ngay góc phải dưới mỗi câu trả lời, giúp người dùng minh bạch về chất lượng dữ liệu.

### 3.3. Thách thức và Giải pháp (Phát triển mới)

**Thách thức:**
- Trong giai đoạn đầu, hệ thống chỉ làm việc được với dữ liệu tĩnh từ file CSV. Điều này khiến chatbot không thể trả lời được các câu hỏi mang tính thời điểm như: *"Hôm nay có những trận đấu nào?"* hay *"Kết quả trận đấu đang diễn ra?"*.
- Nhóm nhận thấy việc thêm thủ công các câu hỏi động này vào database là không khả thi vì dữ liệu thay đổi theo từng giờ.

**Giải pháp: Tích hợp Football API**
- Nhóm quyết định nâng cấp hệ thống lên kiến trúc Hybrid RAG bằng cách gọi trực tiếp dữ liệu từ **Football API**.
- **Ưu điểm:** Cung cấp thông tin chính xác 100% về lịch thi đấu, tỷ số trực tiếp và kết quả thực tế của các giải đấu hàng đầu thế giới.
- **Nhược điểm:** Phụ thuộc vào kết nối internet và giới hạn lượt gọi (Rate limit) của gói API miễn phí.
- **Kết quả:** Việc kết hợp này giúp chatbot giải quyết triệt để vấn đề "dữ liệu chết", mang lại trải nghiệm thực tế và hữu ích cho người hâm mộ bóng đá.

### 3.4. Giao diện người dùng (Premium UI)
- Giao diện được nâng cấp từ phong cách cơ bản sang **Modern Football Premium UI** với các hiệu ứng Glassmorphism, nền ảnh động Messi & Ronaldo, và hệ thống Sidebar chuyên nghiệp.

## 4. SO SÁNH VỚI LÝ THUYẾT ĐÃ HỌC

- **Xác suất Bayes:** Ứng dụng lý thuyết phân lớp xác suất để xử lý bài toán hiểu ý định người dùng (Natural Language Understanding).
- **Instance-based Learning:** Ứng dụng thuật toán KNN và các khái niệm về khoảng cách (Euclidean, Cosine) để giải quyết bài toán tìm kiếm thông tin.
- **Vector Space Model:** Biểu diễn ngôn ngữ tự nhiên dưới dạng các điểm trong không gian đa chiều, cho phép thực hiện các phép toán đại số để so sánh ý nghĩa văn bản.
